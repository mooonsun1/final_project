# 데이터 정제

```python
import os
import shutil
import random
from glob import glob
from zipfile import ZipFile
```

```python
acc = ['S001','S002','S003','S004','S005','S006']
lux = ['L1','L2','L3','L4','L8','L9','L12','L13','L16','L17','L19','L22','L25','L28',]
emotion = ['E01' , 'E02']
angle = ['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9', 'C10', 'C11', 'C12', 'C13', 'C16', 'C19', 'C20']
```

```python
file_path = "/content/drive/MyDrive/data/Low_Resolution/" # 고,중,저 파일이 들어있는 제일 윗 폴더
pure_path = "/content/drive/MyDrive/data/pure_data/" #.
path = os.path.join("/content/drive/MyDrive/data/Low_Resolution","**",)
target_path = os.path.join("/content/drive/MyDrive/data/datasets/")
files = glob(path, recursive=True)
id = []
for file in files:
    name = os.path.basename(file.split('.')[0]) # 파일 이름이 19072121.zip로 되어있어서 .zip를 뺌
    zip_name = os.path.basename(file) # 압축 해제를 하기 위한 파일 이름.
    id.append(name)
del id[0]
file_id = random.sample(range(len(id)),10) # 여기에서 몇개를 고를건지 정하면 됨.
print(file_id)

for file_name in range(len(file_id)):
    zip_file_path = os.path.join(file_path , id[file_id[file_name]]+'.zip')
    zip_file_name = os.path.join(pure_path , id[file_id[file_name]])
    os.makedirs(zip_file_name, exist_ok=True)
    with ZipFile(zip_file_path ,'r') as zip:
        zip.extractall(zip_file_name)
```

```python
counter = 1

for rand in range(len(file_id)):
    rand_id = id[file_id[rand]]
    datasets_name = os.path.join(target_path , rand_id)
    os.makedirs(datasets_name, exist_ok=True)
    a = random.sample(range(len(acc)),2)
    l = random.sample(range(len(lux)),2)
    for i in range(len(emotion)):
        r_acc = acc[a[i]]
        r_lux = lux[l[i]]
        emote = emotion[i]
        for ag in range(len(angle)):
            ang = angle[ag] + '.jpg'
            # print(f"/Users/macbook/Desktop/final/datasets/Low_Resolution/{rand_id}/{r_acc}/{r_lux}/{emote}/{ang}")
            main_file = os.path.join(pure_path,rand_id , r_acc , r_lux , emote , ang,)
            if os.path.exists(main_file):
                new_file_name = f"{rand_id}_{r_acc}_{r_lux}_{emote}_C{counter}.jpg"
                new_file_path = os.path.join(datasets_name , new_file_name)

                shutil.copy2(main_file, new_file_path)
                print(f"Copied {main_file} to {datasets_name}")
            else:
                print(f"File {main_file} does not exist and was skipped.")
```

```python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

train_dataset_path = r"C:\Users\MUNSUNG\Desktop\만선\modeling\train"
valid_dataset_path = r"C:\Users\MUNSUNG\Desktop\만선\modeling\valid"
test_dataset_path = r"C:\Users\MUNSUNG\Desktop\만선\modeling\test"

# 이미지 전처리 과정 정의
transform = transforms.Compose([
    transforms.Resize((160, 160)),  # 모델 입력 크기와 일치하도록 조정
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])  # 정규화
])

# 데이터셋 로드
train_dataset = datasets.ImageFolder(root=train_dataset_path, transform=transform)
valid_dataset = datasets.ImageFolder(root=valid_dataset_path, transform=transform)
test_dataset = datasets.ImageFolder(root=test_dataset_path, transform=transform)

# 데이터 로더 설정
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=64, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
```

```python
from torch.utils.data import Dataset
from PIL import UnidentifiedImageError

class CustomImageFolder(datasets.ImageFolder):
    def __getitem__(self, index):
        try:
            return super(CustomImageFolder, self).__getitem__(index)
        except UnidentifiedImageError:
            print(f"Unidentified image at index {index}, skipping.")
            return None  # None을 반환하여 해당 이미지를 건너뜀

train_dataset = CustomImageFolder(root=train_dataset_path, transform=transform)
valid_dataset = CustomImageFolder(root=valid_dataset_path, transform=transform)
test_dataset = CustomImageFolder(root=test_dataset_path, transform=transform)

# None을 제거하는 새로운 데이터 로더 생성
train_loader = DataLoader([x for x in train_dataset if x is not None], batch_size=64, shuffle=True)
valid_loader = DataLoader([x for x in valid_dataset if x is not None], batch_size=64, shuffle=False)
test_loader = DataLoader([x for x in test_dataset if x is not None], batch_size=64, shuffle=False)
```

```python
import torch
from PIL import Image
from torchvision import transforms

# 추론함수
def predict(image_path, model, transform, device):
    # "model로 image_path의 이미지를 추론한 결과를 반환."
    img = Image.open(image_path)  # 추론대상 이미지 loading
    input_data = transform(img)  # shape: (C, H, W)
    input_data = input_data.unsqueeze(dim=0) # (C, H, W) -> (1, C, H, W)
    input_data = input_data.to(device)

    # 추론
    model = model.to(device)
    model.eval()
    with torch.no_grad():
        pred = model(input_data)
        pred_proba = pred.softmax(dim=-1) # 확률값으로 변경.
        pred_label = pred_proba.argmax(dim=-1).item()  # Tensor([3]) -> 3
        pred_proba_max = f"{pred_proba.max(dim=-1).values.item()*100:.2f}%"
        class_name = train_dataset.classes[pred_label]
        return pred_label, class_name, pred_proba_max
```
