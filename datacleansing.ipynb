```python
import os
import shutil
import random
import splitfolders
from zipfile import ZipFile
from glob import glob
```

```python
acc = ['S001','S002','S003','S004','S005','S006']
lux = ['L1','L2','L3','L4','L8','L9','L12','L13','L16','L17','L19','L22','L25','L28',]
emotion = ['E01' , 'E02']
angle = ['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9', 'C10', 'C11', 'C12', 'C13', 'C16', 'C19', 'C20']
```

```python
low_file_path = "/content/drive/MyDrive/datasets/Low_Resolution/"   # low file 이 저장되어 있는 가장 큰 위치
low_path = " /content/drive/MyDrive/datasets/low_data/"              # 랜덤 추출한 low file 압축 해제 위치
low_zip_path = os.path.join("/content/drive/MyDrive/datasets/Low_Resolution","**",) # id값을 추출하기 위한 경로

middle_file_path = "/content/drive/MyDrive/datasets/Middle_Resolution/"# middle file 이 저장되어 있는 가장 큰 위치
middle_path = "/content/drive/MyDrive/datasets/middle_data/"           # 랜덤 추출한 middle file 압축 해제 위치

high_file_path = "/content/drive/MyDrive/datasets/High_Resolution/"# high file 이 저장되어 있는 가장 큰 위치
high_path = "/content/drive/MyDrive/datasets/high_data/"           # 랜덤 추출한 high file 압축 해제 위치

target_path = os.path.join("/content/drive/MyDrive/datasets/datasets/")  # 랜덤으로 뽑아온 data들 저장할 위치
test_path = os.path.join("/content/drive/MyDrive/datasets/test/")        # test 셋 위치
train_val_path = "/content/drive/MyDrive/datasets"                       # train, val 나누어질 위치
```

```python

################
#파일 id값 불러오기#
################
files = glob(low_zip_path, recursive=True)
id = []
for file in files:
    name = os.path.basename(file.split('.')[0])
    zip_name = os.path.basename(file)
    id.append(name)
del id[0]
file_id = random.sample(range(len(id)),100)  # 랜덤 아이디 추출

r_file_id = random.sample(range(len(file_id)), 50) # 랜덤 아이디 추출한 값에서 테스트 셋을 위해 한번더 랜덤추출
```

```python

##############
#추출 값 압축해제#
##############
for file_name in range(len(file_id)):
    low_zip_file_path = os.path.join(low_file_path , id[file_id[file_name]]+'.zip')
    low_zip_file_name = os.path.join(low_path , id[file_id[file_name]])
    os.makedirs(low_zip_file_name, exist_ok=True)
    with ZipFile(low_zip_file_path ,'r') as low_zip:
        low_zip.extractall(low_zip_file_name)

    middle_zip_file_path = os.path.join(middle_file_path , id[file_id[file_name]]+'.zip')
    middle_zip_file_name = os.path.join(middle_path , id[file_id[file_name]])
    os.makedirs(middle_zip_file_name, exist_ok=True)
    with ZipFile(middle_zip_file_path ,'r') as middle_zip:
        middle_zip.extractall(middle_zip_file_name)


#################
#low train 셋 분류#
#################
for rand in range(len(file_id)):
    rand_id = id[file_id[rand]]
    datasets_name = os.path.join(target_path , rand_id)
    os.makedirs(datasets_name, exist_ok=True)
    a = random.sample(range(len(acc)),2)
    l = random.sample(range(len(lux)),2)
    counter = 1
    for i in range(len(emotion)):
        r_acc = acc[a[i]]
        r_lux = lux[l[i]]
        emote = emotion[i]
        for ag in range(len(angle)):
            ang = angle[ag] + '.jpg'
            # print(f"/Users/macbook/Desktop/final/datasets/Low_Resolution/{rand_id}/{r_acc}/{r_lux}/{emote}/{ang}")
            main_file = os.path.join(low_path, rand_id , r_acc , r_lux , emote , ang,)
            if os.path.exists(main_file):
                new_file_name = f"{rand_id}_{r_acc}_{r_lux}_{emote}_C{counter}.jpg"
                new_file_path = os.path.join(datasets_name, new_file_name)

                # 파일 복사 및 이름 변경
                shutil.copy2(main_file, new_file_path)
                print(f"Copied {main_file} to {new_file_path}")

                # 카운터 증가
                counter += 1
            else:
                print(f"File {main_file} does not exist and was skipped.")
####################
#low train 셋 추가분류#
####################
for add in range(len(file_id)):
    rand_id = id[file_id[add]]   # rand_id = 파일고유아이디
    datasets_name = os.path.join(target_path , rand_id)
    os.makedirs(datasets_name, exist_ok=True)
    counter = 33
    for r in range(0,18):
        a = random.sample(range(len(acc)),1)
        l = random.sample(range(len(lux)),1)
        e = random.sample(range(len(emotion)),1)
        c = random.sample(range(len(angle)),1)
        r_acc = acc[a[0]]
        r_lux = lux[l[0]]
        r_emot = emotion[e[0]]
        r_ang = angle[c[0]] + '.jpg'

        main_file = os.path.join(low_path,rand_id , r_acc , r_lux , r_emot , r_ang,)
        if os.path.exists(main_file):
            new_file_name = f"{rand_id}_{r_acc}_{r_lux}_{emote}_C{counter}.jpg"
            new_file_path = os.path.join(datasets_name, new_file_name)

            shutil.copy2(main_file, new_file_path)
            print(f"Copied {main_file} to {new_file_path}")

            # 카운터 증가
            counter += 1
        else:
            print(f"File {main_file} does not exist and was skipped.")

#################
# low test 셋 분류#
#################

for test in range(len(r_file_id)):
    rand_id = id[file_id[r_file_id[test]]]   # rand_id = 파일고유아이디
    test_datasets_name = os.path.join(test_path , rand_id)
    os.makedirs(test_datasets_name, exist_ok=True)
    counter = 1
    for r in range(0,16):
        a = random.sample(range(len(acc)),1)
        l = random.sample(range(len(lux)),1)
        e = random.sample(range(len(emotion)),1)
        c = random.sample(range(len(angle)),1)
        r_acc = acc[a[0]]
        r_lux = lux[l[0]]
        r_emot = emotion[e[0]]
        r_ang = angle[c[0]] + '.jpg'

        main_file = os.path.join(low_path,rand_id , r_acc , r_lux , r_emot , r_ang,)
        if os.path.exists(main_file):
            new_file_name = f"{rand_id}_{r_acc}_{r_lux}_{emote}_C{counter}.jpg"
            new_file_path = os.path.join(test_datasets_name, new_file_name)

            shutil.copy2(main_file, new_file_path)
            print(f"Copied {main_file} to {new_file_path}")

            # 카운터 증가
            counter += 1
        else:
            print(f"File {main_file} does not exist and was skipped.")

######################################################################################################
####################################### Middle 데이터 셋 분류 ############################################
######################################################################################################
####################
#middle train 셋 분류#
####################
for rand in range(len(file_id)):
    rand_id = id[file_id[rand]]
    datasets_name = os.path.join(target_path , rand_id)
    os.makedirs(datasets_name, exist_ok=True)
    a = random.sample(range(len(acc)),2)
    l = random.sample(range(len(lux)),2)
    counter = 51

    for i in range(len(emotion)):
        r_acc = acc[a[i]]
        r_lux = lux[l[i]]
        emote = emotion[i]

        for ag in range(len(angle)):
            ang = angle[ag] + '.jpg'
            main_file = os.path.join(middle_path, rand_id , r_acc , r_lux , emote , ang,)

            if os.path.exists(main_file):
                new_file_name = f"{rand_id}_{r_acc}_{r_lux}_{emote}_C{counter}.jpg"
                new_file_path = os.path.join(datasets_name, new_file_name)

                # 파일 복사 및 이름 변경
                shutil.copy2(main_file, new_file_path)
                print(f"Copied {main_file} to {new_file_path}")

                # 카운터 증가
                counter += 1
            else:
                print(f"File {main_file} does not exist and was skipped.")
#######################
#middle train 셋 추가분류#
#######################
for add in range(len(file_id)):
    rand_id = id[file_id[add]]   # rand_id = 파일고유아이디
    datasets_name = os.path.join(target_path , rand_id)
    os.makedirs(datasets_name, exist_ok=True)
    counter = 83
    for r in range(0,18):
        a = random.sample(range(len(acc)),1)
        l = random.sample(range(len(lux)),1)
        e = random.sample(range(len(emotion)),1)
        c = random.sample(range(len(angle)),1)

        r_acc = acc[a[0]]
        r_lux = lux[l[0]]
        r_emot = emotion[e[0]]
        r_ang = angle[c[0]] + '.jpg'

        main_file = os.path.join(middle_path,rand_id , r_acc , r_lux , r_emot , r_ang,)

        if os.path.exists(main_file):
            new_file_name = f"{rand_id}_{r_acc}_{r_lux}_{emote}_C{counter}.jpg"
            new_file_path = os.path.join(datasets_name, new_file_name)

            shutil.copy2(main_file, new_file_path)
            print(f"Copied {main_file} to {new_file_path}")

            # 카운터 증가
            counter += 1
        else:
            print(f"File {main_file} does not exist and was skipped.")

####################
# middle test 셋 분류#
####################

for test in range(len(r_file_id)):
    rand_id = id[file_id[r_file_id[test]]]   # rand_id = 파일고유아이디
    test_datasets_name = os.path.join(test_path , rand_id)
    os.makedirs(test_datasets_name, exist_ok=True)
    counter = 17
    for r in range(0,16):
        a = random.sample(range(len(acc)),1)
        l = random.sample(range(len(lux)),1)
        e = random.sample(range(len(emotion)),1)
        c = random.sample(range(len(angle)),1)
        r_acc = acc[a[0]]
        r_lux = lux[l[0]]
        r_emot = emotion[e[0]]
        r_ang = angle[c[0]] + '.jpg'

        main_file = os.path.join(middle_path,rand_id , r_acc , r_lux , r_emot , r_ang,)
        if os.path.exists(main_file):
            new_file_name = f"{rand_id}_{r_acc}_{r_lux}_{emote}_C{counter}.jpg"
            new_file_path = os.path.join(test_datasets_name, new_file_name)

            shutil.copy2(main_file, new_file_path)
            print(f"Copied {main_file} to {new_file_path}")

            # 카운터 증가
            counter += 1
        else:
            print(f"File {main_file} does not exist and was skipped.")

######################################################################################################
####################################### High 데이터 셋 분류 ############################################
######################################################################################################
####################
#high train 셋 분류#
####################
for rand in range(len(file_id)):
    rand_id = id[file_id[rand]]
    datasets_name = os.path.join(target_path , rand_id)
    os.makedirs(datasets_name, exist_ok=True)
    a = random.sample(range(len(acc)),2)
    l = random.sample(range(len(lux)),2)
    counter = 101

    for i in range(len(emotion)):
        r_acc = acc[a[i]]
        r_lux = lux[l[i]]
        emote = emotion[i]

        for ag in range(len(angle)):
            ang = angle[ag] + '.jpg'
            main_file = os.path.join(high_path, rand_id , r_acc , r_lux , emote , ang,)

            if os.path.exists(main_file):
                new_file_name = f"{rand_id}_{r_acc}_{r_lux}_{emote}_C{counter}.jpg"
                new_file_path = os.path.join(datasets_name, new_file_name)

                # 파일 복사 및 이름 변경
                shutil.copy2(main_file, new_file_path)
                print(f"Copied {main_file} to {new_file_path}")

                # 카운터 증가
                counter += 1
            else:
                print(f"File {main_file} does not exist and was skipped.")
#######################
#high train 셋 추가분류#
#######################
for add in range(len(file_id)):
    rand_id = id[file_id[add]]   # rand_id = 파일고유아이디
    datasets_name = os.path.join(target_path , rand_id)
    os.makedirs(datasets_name, exist_ok=True)
    counter = 133
    for r in range(0,18):
        a = random.sample(range(len(acc)),1)
        l = random.sample(range(len(lux)),1)
        e = random.sample(range(len(emotion)),1)
        c = random.sample(range(len(angle)),1)

        r_acc = acc[a[0]]
        r_lux = lux[l[0]]
        r_emot = emotion[e[0]]
        r_ang = angle[c[0]] + '.jpg'

        main_file = os.path.join(high_path,rand_id , r_acc , r_lux , r_emot , r_ang,)

        if os.path.exists(main_file):
            new_file_name = f"{rand_id}_{r_acc}_{r_lux}_{emote}_C{counter}.jpg"
            new_file_path = os.path.join(datasets_name, new_file_name)

            shutil.copy2(main_file, new_file_path)
            print(f"Copied {main_file} to {new_file_path}")

            # 카운터 증가
            counter += 1
        else:
            print(f"File {main_file} does not exist and was skipped.")

####################
# high test 셋 분류#
####################

for test in range(len(r_file_id)):
    rand_id = id[file_id[r_file_id[test]]]   # rand_id = 파일고유아이디
    test_datasets_name = os.path.join(test_path , rand_id)
    os.makedirs(test_datasets_name, exist_ok=True)
    counter = 33
    for r in range(0,17):
        a = random.sample(range(len(acc)),1)
        l = random.sample(range(len(lux)),1)
        e = random.sample(range(len(emotion)),1)
        c = random.sample(range(len(angle)),1)
        r_acc = acc[a[0]]
        r_lux = lux[l[0]]
        r_emot = emotion[e[0]]
        r_ang = angle[c[0]] + '.jpg'

        main_file = os.path.join(high_path,rand_id , r_acc , r_lux , r_emot , r_ang,)
        if os.path.exists(main_file):
            new_file_name = f"{rand_id}_{r_acc}_{r_lux}_{emote}_C{counter}.jpg"
            new_file_path = os.path.join(test_datasets_name, new_file_name)

            shutil.copy2(main_file, new_file_path)
            print(f"Copied {main_file} to {new_file_path}")

            # 카운터 증가
            counter += 1
        else:
            print(f"File {main_file} does not exist and was skipped.")
```
